```python
import sys
import os
import pandas as pd

sys.path.append(os.path.abspath(".."))  
from models.Naive_Bayes import GaussianNaiveBayes, accuracy_score

# Path to the CSV file
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))  
csv_path = os.path.join(project_root, "..", "data", "IRIS_cleaned.csv") 
feature_cols = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# Demo Gaussian Naive Bayes
gnb = GaussianNaiveBayes()
df = pd.read_csv(csv_path)
df.columns = df.columns.str.strip()  # Loại bỏ khoảng trắng trong tên cột

X = df[feature_cols].values
y = df["species"].values
gnb.fit(X, y)
y_pred = gnb.predict(X)
acc = accuracy_score(y, y_pred)
print("\n=== GAUSSIAN NAIVE BAYES ===") 
print(f"Accuracy on entire dataset: {acc:.4f}")
```
